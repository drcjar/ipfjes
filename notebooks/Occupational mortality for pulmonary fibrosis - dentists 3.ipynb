```python
import pandas as pd
```

```python
df = pd.read_csv('../data/Men16_64.csv')
# Occupational mortality in England and Wales, 1991-2000 - Men age 65-74
# from http://webarchive.nationalarchives.gov.uk/20150907090421/http://www.ons.gov.uk/ons/publications/re-reference-tables.html?edition=tcm%3A77-168405
```
