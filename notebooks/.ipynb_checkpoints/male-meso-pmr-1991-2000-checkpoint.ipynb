```python
import xlrd
import pandas as pd
```

```python
df = pd.read_excel('../data/occ80007.xls', skiprows=2) 

# Mesothelioma proportional mortality ratios (PMRs) for males aged 16-74 in 
# Great Britain by Standard Occupational Classification (1990) unit group, 1991-2000 
```

```python
df.head()
```

```python
print('There are {0} jobs for which the lower confidence interval is > 100'.format(len(df[df.Lower > 100])))
```

```python
df[df.Lower > 100].sort_values(by='PMR', ascending=False)

## SOC code, occupation etc for significant PMRs (95% CI does not include 100)
```

```python
print('There are {0} jobs for which there are at least 30 deaths and the' \
      'lower confidence interval is > 100'.format(len(df[(df.Lower > 100) & (df.Deaths > 30)])))
```

```python
df[(df.Lower > 100) & (df.Deaths > 30)].sort_values(by='PMR', ascending=False)
```

```python
print('There are {0} jobs for which there are at least 30 deaths and the' \
      'lower confidence interval is > 100'.format(len(df[(df.Lower > 100) & (df.Deaths > 30) & (df.PMR > 200)])))
```

```python
df[(df.Lower > 100) & (df.Deaths > 30) & (df.PMR > 200)]
# This is the same as Peto's Table 2.3.1 in http://www.hse.gov.uk/research/rrpdf/rr696.pdf
```
